```python
from PIL import Image
import numpy as np
 
im = Image.open('boat.png').convert('L')
im_array = np.array(im)
im_inverse_array = 255 - im_array
im_inverse = Image.fromarray(im_inverse_array)
im_inverse.save('result.png')
im_inverse
```

The code takes the the matrix of the image and then inverses the colors and then converting it to the inversed picture and saving it locally.

```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

im = Image.open('boat.png').convert('L')
im = np.array(im)/255.0  # convert image to a numpy array and perform division
im_power_law_transformation = cv2.pow(im, 0.6)

plt.imshow(im, cmap='gray')  # display the original image using matplotlib
plt.show()

plt.imshow(im_power_law_transformation, cmap='gray')  # display the transformed image using matplotlib
plt.show()
```
